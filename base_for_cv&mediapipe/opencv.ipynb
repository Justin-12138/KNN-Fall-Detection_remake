```python
# 读取视频并显示帧率
import time

# 创建一个 VideoCapture 对象
cap = cv2.VideoCapture(0)

# 获取视频的宽度和高度
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
# 初始化时间和帧数
prevTime = 0
# 使用无限循环来持续获取摄像头的画面
while True:
    # 读取一帧画面
    ret, frame = cap.read()
    # 如果读取成功，显示这帧画面
    if ret:
        # 计算帧率
        currTime = time.time()
        fps = 1 / (currTime - prevTime)
        prevTime = currTime
        # 在画面上添加帧率和窗口大小
        cv2.putText(frame, f'FPS: {fps:.2f}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
        cv2.putText(frame, f'Size: {width}x{height}', (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
        # 显示画面
        cv2.imshow('frame', frame)
    # 如果按下 'q' 键，就退出循环
    key = cv2.waitKey(1) & 0xFF
    if key == ord('s'):
        cv2.imwrite("key_frame.jpg" ,frame)
    else:
        break
    
# 释放摄像头
cap.release()
# 关闭所有窗口
cv2.destroyAllWindows()
```
